### List 1 - Tweet

### importing function to tokenize data in the tweets


```python
from nltk.tokenize import word_tokenize              
```

### storing the tokenized words in tokens   

```python
tokens = (word_tokenize(tweets))                     
```

```python
type(tokens)
```

```python
tokens[:20]
```

### checking if tokenized word is alphabet otherwise discaring it

```python
tweet_words = [w.lower() for w in tokens if w.isalpha()]
```

```python
len(tweet_words)
tweet_words[:10]
```

### Use Dictionary instead of List to improve the effieiency

### Dictionary 1 - Positive 

```python
positive = {}
f = open('positive.txt', 'r')
for w in f:
    w = w.strip()
    positive[w] = 0
len(positive)
#positive
f.close()
```

### Dictionary 2 - Negative

```python
negative = {}
f = open('negative.txt', 'r')
for w in f:
    w = w.strip()
    negative[w] = 0
len(negative)
f.close()
```

### Dictionary 3 - Stopword

```python
stopword = {}
f = open('stopwords.txt', 'r')
for w in f:
    w = w.strip()
    stopword[w] = 0
len(stopword)
f.close()
```

### Compare the words in tweet with 3 Dictionaries - positive, negative and stopword


A dictionary is a lookup, while a list is an iteration. Dictionary uses a **hash lookup**, while your list requires walking through the list until it finds the result from beginning to the result each time.

```python
positive['a+']
```

```python
count_s = 0
count_p = 0
count_n = 0
count_g = 0

for w in tweet_words:
    if (w in positive) & (w !='trump'):
        count_p += 1
        positive[w] +=1
    elif w in negative:
        count_n += 1
    elif w in stopword:
        count_s += 1
    else:
        count_g += 1
print ("positive word count = %d " % count_p)
print ("negative word count = %d " % count_n)
print ("stopword word count = %d " % count_s)
print ("general word count = %d " % count_g)
print ("total word count = %d " % (count_p + count_n + count_s + count_g))
print ('==========================================')
print ('positive vs nagative ratio = %.2f ' % (count_p/count_n))

```

### Extra note: Hashing / Hash Table

*index = f(key, array_size)*

Often this is done in two steps:

*hash = hashfunc(key)*

*index = hash % array_size*
